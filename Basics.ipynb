```python
%%sql



SELECT MIN(population) AS min_population,
       MAX(population) AS max_population,
       MIN(population_growth) AS min_poplatn_grth,
       MAX(population_growth) AS  max_poplatn_grth
FROM   facts

WHERE name <> 'World'

```

```python

%%sql

SELECT AVG(population) AS avg_population, 
       AVG(area) AS avg_area
    
FROM   facts

WHERE  name <> 'World';
```

```python
%%sql

SELECT *

FROM  facts

WHERE population > (SELECT AVG(population) 
                    FROM   facts
                    WHERE  name <> 'World') 

      AND   area <  (SELECT AVG(area) 
                     FROM   facts
                     WHERE  name <> 'World')
;
```

```python
%%sql


SELECT name AS most_populated,
       MAX(Population)
    
FROM   facts

WHERE name <> 'World'
```

```python
%%sql


SELECT name AS highest_growth_rate,
       MAX(population_growth)
    
FROM   facts

WHERE name <> 'World'
```

```python
%%sql

SELECT AVG(CAST(area_water AS Float) / area_land) AS 
                                avg_water_to_land
FROM facts
WHERE  name <> 'World'
```

```python
%%sql

SELECT *, CAST(area_water AS Float) / area_land AS 
       water_to_land

FROM   facts

WHERE  water_to_land > (SELECT AVG(CAST(area_water AS Float) / area_land) AS 
                                avg_water_to_land
                                FROM facts
                                WHERE  name <> 'World')
    

```

```python
%%sql

SELECT *

FROM  facts

WHERE area_water > area_land AND name <> 'World';
```

```python
%%sql

SELECT *
FROM   facts
WHERE birth_rate >  (SELECT AVG(birth_rate)
                     FROM   facts
                     WHERE name <> 'World')

ORDER BY birth_rate DESC



```

```python
%%sql

SELECT name, death_rate, birth_rate

FROM facts

WHERE death_rate > birth_rate AND name <> 'World'
```
